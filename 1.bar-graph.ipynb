```r
#載入ggplot2和workspace
library(ggplot2)
#load("C:/Users/6609/Documents/data.rData")
Sample <- read.csv("data/Sample.csv")
```

```r
head(Sample)
```

```r
str(Sample)
```

```r
#載入plyr,以state為raw將profit進行加總
library(plyr)
sample_bar <- ddply(Sample, ~State.or.Province, summarise, profit = sum(Profit))
```

```r
head(sample_bar)
```

```r
#bar chart
bar_chart <- ggplot(sample_bar,aes(x=State.or.Province,y=profit))+geom_bar(stat = "identity")
```

```r
bar_chart
```

```r
#轉置x/y-axis
bar_chart+coord_flip()
```

```r
#排序y-axis
bar_chart + coord_flip() + scale_x_discrete(limits=rev(levels(Sample$State.or.Province)))
```

```r
bar_chart + coord_flip() + scale_x_discrete(limits=rev(levels(Sample$State.or.Province))) + scale_y_continuous(breaks=seq(-20000, 1300000, 20000))
```
